```python
#Importing data with yfinance as yahoo_fin didn't work
GOOG= yf.download("GOOG", start="2019-01-01", end="2019-03-30")
GOOG = pd.DataFrame(data=GOOG)
GOOG.columns=['Open','High','Low','Close','Adjclose','Volume']
GOOG.reset_index(level=0, inplace=True)
GOOG
```

```python
GOOG['Date'] = pd.to_datetime(GOOG['Date'])
```

```python
GOOG = GOOG.sort_values(by=['Date'])
```

```python
#Changing the datatype 
GOOG.Open = GOOG.Open.astype(float)
GOOG.Close = GOOG.Close.astype(float)
GOOG.High = GOOG.High.astype(float)
GOOG.Low = GOOG.Low.astype(float)
GOOG.Volume = GOOG.Volume.astype(int)
GOOG.Adjclose = GOOG.Adjclose.astype(int)
#check the data
GOOG.head()
```

```python
GOOG.info()
```

```python
#USING BOKEH TO VISUALIZE THE STOCKS OF GOOGLE FOR THE SELECTED PERIOD
inc = GOOG.Close > GOOG.Open
dec = GOOG.Open > GOOG.Close
w = 12*60*60*1000 # half day in ms
TOOLS = "pan,wheel_zoom,box_zoom,reset,save"
title = "GOOG" + ' Chart'
p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=1000, title = title)
p.xaxis.major_label_orientation = pi/4
p.grid.grid_line_alpha=0.3
p.segment(GOOG.Date, GOOG.High, GOOG.Date, GOOG.Low, color="black")
p.vbar(GOOG.Date[inc], w, GOOG.Open[inc], GOOG.Close[inc], fill_color="#D5E1DD", line_color="black")
p.vbar(GOOG.Date[dec], w, GOOG.Open[dec], GOOG.Close[dec], fill_color="#F2583E", line_color="black")
#Store as a HTML file
webbrowser.open_new_tab("stock_information.html")
# Display in browser
show(p)
```

```python
GOOG1 = GOOG[['Adjclose']] 
# Take a look at the new data 
print(GOOG1.head())
```

```python
# A variable for predicting 'n' days out into the future
forecast_out4 = 30 #'n=30' days
#Create another column (the target ) shifted 'n' units up
GOOG1['Prediction'] = GOOG1[['Adjclose']].shift(-forecast_out4)
#print the new data set
print(GOOG1.tail())
```

```python
### Create the independent data set (X)  #######
# Convert the dataframe to a numpy array
X4 = np.array(GOOG1.drop(['Prediction'],1))

#Remove the last '30' rows
X4 = X4[:-forecast_out4]
print(X4)
```

```python
### Create the dependent data set (y)  #####
# Convert the dataframe to a numpy array 
y4 = np.array(GOOG1['Prediction'])
# Get all of the y values except the last '30' rows
y4 = y4[:-forecast_out4]
print(y4)
```

```python
# Split the data into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(X4, y4, test_size=0.2)
```

```python
# Create and train the Linear Regression  Model
lr = LinearRegression()
# Train the model
lr.fit(x_train, y_train)
```

```python
lr_confidence4 = lr.score(x_test, y_test)
print("lr confidence: ", lr_confidence4)
```

```python
x_forecast4 = np.array(GOOG1.drop(['Prediction'],1))[-forecast_out4:]
print(x_forecast4)
```

```python
# PREDICTING FUTURE STOCKS FOR NEXT 30 DAYS
lr_prediction4 = lr.predict(x_forecast4)
print(lr_prediction4)
```

```python
#DEFINING A DATE RANGE FTOM APRIL 1 2019 TO APRIL 30 2019.
lr_date = pd.date_range(start='01/04/2019',end= '30/04/2019')
lr_date
```

```python
#PLOTTING THE PREDICTIONS ON A LINE GRAPH 
from bokeh.plotting import figure, output_notebook, show
output_notebook()
p4 = figure(plot_width = 400, plot_height = 400) 
p4.line(lr_date,lr_prediction4,
        line_width = 2, color = "green") 
show(p4)
```
